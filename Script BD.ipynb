# [dbo].[sp_Login]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Login]    Script Date: 2/03/2024 10:44:33 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[sp_Login]

@Usuario varchar(50),
@Clave varchar(max),
@Patron varchar(50)
as begin

select * from Usuarios where Usuario=@Usuario and convert(varchar(50), DECRYPTBYPASSPHRASE(@Patron, Clave)) = @Clave
end
GO
```

# [dbo].[sp_Permisos]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Permisos]    Script Date: 2/03/2024 10:44:33 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

create procedure [dbo].[sp_Permisos]

@Id_Rol int
as begin

select Nombre_Permiso, Estado from Roles_Permisos inner join Permisos on Permisos.Id_Permiso = Roles_Permisos.Id_Permiso where Id_Rol = @Id_Rol
end
GO
```

# [dbo].[sp_registrar]

```sql
/****** Object:  StoredProcedure [dbo].[sp_registrar]    Script Date: 2/03/2024 10:44:33 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[sp_registrar]
@Nombre varchar(50),
@Edad int,
@Usuario varchar(50),
@Clave varchar(max),
@Id_Rol int = 2,
@Patron varchar(50)

as begin

insert into Usuarios values (@Nombre, @Edad, @Usuario, ENCRYPTBYPASSPHRASE(@Patron, @Clave), @Id_Rol)

end
GO
```
